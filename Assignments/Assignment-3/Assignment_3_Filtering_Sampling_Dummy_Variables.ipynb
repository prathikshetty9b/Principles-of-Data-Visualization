### Detecting and Filtering Outliers

Filtering  or  transforming  outliers  is  largely  a  matter  of  applying  array  operations.

```python
import pandas as pd
import numpy as np

#create a dataframe
data = pd.DataFrame(np.random.randn(1000,4))

data.describe()
```

```python
#find  values  in  one  of  the  columns  exceeding  3  in  absolutevalue
col = data[2]
col[np.abs(col) > 3]
```

```python
#To select all rows having a value exceeding 3 or –3, you can use the any method on aboolean DataFrame:
data[(np.abs(data) > 3).any(1)]
```

```python
#Values can be set based on these criteria. Here is code to cap values outside the inter‐val –3 to 3:
data[np.abs(data) > 3] = np.sign(data) * 3
data.describe()
```

```python
#np.sign(data) produces 1 and –1 values based on whether the valuesin data are positive or negative:
np.sign(data).head()
```

### Computing Indicator/Dummy Variables

Another type of transformation for statistical modeling or machine learning applications  is  converting  a  categorical  variable  into  a  “dummy”  or  “indicator”  matrix.  If  acolumn  in  a  DataFrame  has  k  distinct  values,  you  would  derive  a  matrix  or  Data‐Frame  with  k  columns  containing  all  1s  and  0s.  pandas  has  a  get_dummies  functionfor  doing  this,  though  devising  one  yourself  is  not  difficult. 

```python
df = pd.DataFrame({'key':['b','b','a','c','a','b'],'data1':range(6)})
pd.get_dummies(df['key'])
```

```python
#In  some  cases,  you  may  want  to  add  a  prefix  to  the  columns  in  the  indicator  Data‐Frame, which can then be merged with the other data. get_dummies has a prefix argument for doing this:
dummies = pd.get_dummies(df['key'], prefix='key')
df_with_dummy = df[['data1']].join(dummies)
df_with_dummy
```

### Movies Genres Example from Computing Indicator/Dummy Variables

```python
mnames = ['movie_id', 'title', 'genres']
```

```python
movies = pd.read_table('/content/movies.dat', sep='::',header=None, names=mnames)
movies[:10]
```

```python
#Adding  indicator  variables  for  each  genre  requires  a  little  bit  of  wrangling.  First,  we extract the list of unique genres in the dataset:
all_genres = []
for x in movies.genres:
  all_genres.extend(x.split('|'))

genres = pd.unique(all_genres)
genres
```

```python
#One  way  to  construct  the  indicator  DataFrame  is  to  start  with  a  DataFrame  of  allzeros:
zero_matrix = np.zeros((len(movies), len(genres)))
#Converting matrix to dataFrame
dummies = pd.DataFrame(zero_matrix, columns=genres)
dummies
```

```python
gen = movies.genres[0]
gen.split('|')
```

```python
dummies.columns.get_indexer(gen.split('|'))
```

```python
#Now,  iterate  through  each  movie  and  set  entries  in  each  row  of  dummies  to  1.
for i, gen in enumerate(movies.genres):
  indices = dummies.columns.get_indexer(gen.split('|'))
  dummies.iloc[i, indices] = 1
dummies
```

```python
#combine
movies_windic = movies.join(dummies.add_prefix('Genre_'))
movies_windic.iloc[0:2:]
```

```python
#A useful recipe for statistical applications is to combine get_dummies with a discreti‐zation function like cut:
np.random.seed(12345)
values = np.random.rand(10)
values
```

```python
bins = [0, 0.2, 0.4, 0.6, 0.8, 1]
pd.get_dummies(pd.cut(values, bins))
```
